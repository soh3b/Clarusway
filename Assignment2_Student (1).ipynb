** Calculate average of spending according to country with apply function? **

```python
def avgspend(n):
    count = n.count()
    tota = n.sum()
    avg = tota/count
    return avg

cont1 = ecom.groupby('Country')['UnitPrice']
cont1.apply(avgspend)
```

** Calculate average,mean,max of spending according to country with agg function? **

```python
cont1 = ecom.groupby('Country')['UnitPrice'].agg({'A' : ['sum', 'min'], 'B' : ['min', 'max']})
cont1
```

** How many people who live in Spain spend more than average Total ?**

```python
spains = ecom.where(ecom['Country'] == 'Spain')
truespain = spains['Total']
avg = ecom['Total'].mean()

abvavg = truespain >avg
abvavg.value_counts()
print("728 people in Spain spend more than the average Total")
```
